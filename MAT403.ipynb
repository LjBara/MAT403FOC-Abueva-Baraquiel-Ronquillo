```python
import math

def measured_fps(p):

    return 120.0 - 0.1109 * (p ** 1.4776)

def performance_function(p, target_fps=60.0):

    return measured_fps(p) - target_fps

def bisection_method(a, b, target_fps=60.0, tol=0.3, max_iter=100):
    fa = performance_function(a, target_fps)
    fb = performance_function(b, target_fps)

    if fa * fb > 0:
        raise ValueError(
            'Sign condition not met. Check that the target FPS '
            'is achievable somewhere within [a, b].'
        )

    print(f"{'Iter':>4}  {'a':>8}  {'b':>8}  {'m':>10}  "
          f"{'FPS(m)':>9}  {'f(m)':>8}")
    print('-' * 60)

    for n in range(1, max_iter + 1):
        m = (a + b) / 2.0
        fps_m = measured_fps(m)
        fm = performance_function(m, target_fps)

        print(f'{n:>4}  {a:>8.4f}  {b:>8.4f}  {m:>10.4f}  '
              f'{fps_m:>9.4f}  {fm:>+8.4f}')

        if abs(fm) < tol:
            print(f'\nConverged at iteration {n}.')
            print(f'Optimal setting: p* = {m:.4f}')
            print(f'FPS at p*: {fps_m:.4f}')
            return m

        if fa * fm < 0:
            b, fb = m, fm
        else:
            a, fa = m, fm

    return (a + b) / 2.0

if __name__ == '__main__':
    print('Sign check:')
    print(f'  f(0)   = {performance_function(0):.4f}  -> above target')
    print(f'  f(100) = {performance_function(100):.4f} -> below target')
    print()
    optimal = bisection_method(0, 100, target_fps=60, tol=0.3)
    print(f'\nFinal answer: Set quality to {optimal:.1f} out of 100')
    print(f'Theoretical bound: {math.ceil(math.log2(100/0.3))} iterations')
```
